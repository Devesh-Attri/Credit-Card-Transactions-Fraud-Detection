#**1️⃣ Load the Data**

```python
import pandas as pd

# Load the dataset
train_path = "/content/drive/MyDrive/Credit Card Transactions Fraud Detection/fraudTrain.csv"  # Update with the correct path
test_path = "/content/drive/MyDrive/Credit Card Transactions Fraud Detection/fraudTest.csv" # Update with the correct path

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# Display basic information about the dataset
print(df_train.info())
df_train.head()  # View first few rows
```

```python
print(df_test.info())
df_test.head()
```

#**2️⃣ Preprocess Data**

###**Handle Date/Time Features**

```python
# Check for missing values
print(df_train.isnull().sum())
print(df_test.isnull().sum())
```

###**Handle Age Feature**

```python
# Convert transaction timestamp to datetime format
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])

# Extract relevant time-based features
for df in [df_train, df_test]:
    df['hour'] = df['trans_date_trans_time'].dt.hour
    df['day'] = df['trans_date_trans_time'].dt.day
    df['month'] = df['trans_date_trans_time'].dt.month
    df['year'] = df['trans_date_trans_time'].dt.year
```

```python
from datetime import datetime

# Convert 'dob' to age
def calculate_age(dob):
    today = datetime.today()
    birth_date = datetime.strptime(dob, "%Y-%m-%d")
    return today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))

df_train['age'] = df_train['dob'].apply(calculate_age)
df_test['age'] = df_test['dob'].apply(calculate_age)
```

#**3️⃣ Encode Categorical Variables**

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['category', 'gender', 'job', 'state']

encoders = {}  # Dictionary to store encoders for each column

for col in categorical_cols:
    encoders[col] = LabelEncoder()

    # Fit on training data
    df_train[col] = encoders[col].fit_transform(df_train[col])

    # Transform test data (handle unseen categories)
    df_test[col] = df_test[col].apply(lambda x: encoders[col].transform([x])[0] if x in encoders[col].classes_ else -1)
```

#**4️⃣ Select Features for Model Training**

```python
# Define features to use
features = ['amt', 'hour', 'day', 'month', 'year', 'category', 'lat', 'long', 'city_pop', 'job', 'age']
target = 'is_fraud'

# Prepare training and test data
X_train, y_train = df_train[features], df_train[target]
X_test, y_test = df_test[features], df_test[target]
```

#**5️⃣ Train a Machine Learning Model**

```python
from re import VERBOSE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Split training data into train/validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Train Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate on validation set
y_val_pred = model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

#**6️⃣ Test Model on Unseen Data**

```python
# Predict on test data
y_test_pred = model.predict(X_test)

# Evaluate model performance
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

#**7️⃣ Analyze Misclassifications**

```python
df_test['predicted'] = y_test_pred
df_test['actual'] = y_test

# Find misclassified transactions
misclassified = df_test[df_test['predicted'] != df_test['actual']]
print("Misclassified Transactions:\n", misclassified.head())
```
